## Tutorial for IPrivPy

### Query Data Source List

% query [#page] [page_size]

For example, "% query 1 10" means show me the first page which contains no more than 10 entries each page.

Note: the command must be separated by a space.

```python
% query 2 10
```

### Bind Variables with the Data Source
% bind

[task_name]

-i [input_var1_name] [dataSourceMetadataId] [datasourceId] [company1_id]

-i [input_var2_name] [dataSourceMetadataId] [datasourceId] [company1_id]

-o [output_var1_name]  [company2_id]

-o [output_var2_name]  [company2_id]

......

Note: every line of the command above must be separated by a space.

So far, we only support one result_name.

| param       	| description                        	|
|-------------	|------------------------------------	|
| task_name   	| task name related with the session 	|
| output_var1_name 	| used in the command `print`        	|
| input_var1_name    	| variable name                      	|
| DS_id       	| Data Source ID                     	|
| company_id  	| company ID                         	|

```python
% bind
sherwin.yan-test
-i a1  69 170 3
-i a2  68 170 3
-o a3 -1
```
